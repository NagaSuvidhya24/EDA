```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets from CSV files
transactions = pd.read_csv('Transactions.csv')
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products (1).csv')

# Clean column names
transactions.columns = transactions.columns.str.strip()
customers.columns = customers.columns.str.strip()
products.columns = products.columns.str.strip()
```

```python
# Merge the datasets on CustomerID and ProductID
merged_df = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')

# Convert TransactionDate to datetime
merged_df['TransactionDate'] = pd.to_datetime(merged_df['TransactionDate'])
```

```python
# Check for missing values
print("Missing Values:\n", merged_df.isnull().sum())

# Check data types
print("Data Types:\n", merged_df.dtypes)

# Basic statistics
print("Summary Statistics:\n", merged_df.describe())
```

```python
# Visualize Quantity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['Quantity'], bins=10, kde=True)
plt.title('Quantity Distribution')
plt.show()

# Visualize Revenue Distribution
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['TotalValue'], bins=10, kde=True, color='orange')
plt.title('Revenue Distribution')
plt.show()
```

```python
# Add Month-Year column
merged_df['MonthYear'] = merged_df['TransactionDate'].dt.to_period('M')

# Group by MonthYear to calculate total revenue
monthly_sales = merged_df.groupby('MonthYear')['TotalValue'].sum()

# Plot sales trends
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Sales Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
# Top 5 products by revenue
top_products = merged_df.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head(5)

# Plot top products
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar', color='green')
plt.title('Top 5 Products by Revenue')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Customer distribution by region
region_distribution = merged_df['Region'].value_counts(normalize=True) * 100

# Plot region distribution
plt.figure(figsize=(8, 6))
region_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Customer Distribution by Region')
plt.ylabel('')  # Hide y-label for pie chart
plt.show()
```

```python
# Calculate repeat customer rate
repeat_customers = merged_df['CustomerID'].value_counts()
repeat_rate = (repeat_customers[repeat_customers > 1].count() / repeat_customers.count()) * 100
print(f"Repeat Customer Rate: {repeat_rate:.2f}%")
```
